# U.S. Medical Insurance Costs

```python

import csv
import numpy as np
import matplotlib.pyplot as plt

insurance_cost_by_age = {}
avg_insurance_cost_by_age = {}

with open('insurance.csv', 'r', newline='') as insurance_dataset:
    dataset_dict = csv.DictReader(insurance_dataset)
    
    for row in dataset_dict:
        age = row.get('age')
        insurance_cost = row.get('charges')
        if age not in insurance_cost_by_age:
            insurance_cost_by_age[age] = [float(insurance_cost)]
        else:
            insurance_cost_by_age[age].append(float(insurance_cost))
            
for age in sorted(insurance_cost_by_age.keys()):
    avg_insurance_cost_by_age[age] = round(sum(insurance_cost_by_age[age]) / len(insurance_cost_by_age[age]),2)
    
```

```python

#setting up plot
ax = plt.subplot()
ax.set_xticks([i for i in range(0,50,3)])


age = [age for age in avg_insurance_cost_by_age]
avg_cost = [cost for cost in avg_insurance_cost_by_age.values()]

ax.plot(age, avg_cost)

plt.title('Average Insurance Cost Based on Age for Men and Women')
plt.xlabel('Age')
plt.ylabel('Cost in dollars')

ax.axis([0,45,3000,27000])
plt.show()
```

```python

insurance_costs_men = {}
avg_insurance_costs_men = {}
insurance_costs_women = {}
avg_insurance_costs_women = {}

with open('insurance.csv', 'r', newline='') as insurance_dataset:
    dataset_dict = csv.DictReader(insurance_dataset)
    
    for row in dataset_dict:
        age = row.get('age')
        sex = row.get('sex')
        cost = float(row.get('charges'))
        
        if sex == 'male':
            if age in insurance_costs_men:
                insurance_costs_men[age].append(cost)
            else:
                insurance_costs_men[age] = [cost]
                
        elif sex == 'female':
            if age in insurance_costs_women:
                insurance_costs_women[age].append(cost)
            else:
                insurance_costs_women[age] = [cost]
                
for age in sorted(insurance_costs_men.keys()):
    avg_insurance_costs_men[age] = round(sum(insurance_costs_men[age]) / len(insurance_costs_men[age]),2)

for age in sorted(insurance_costs_women.keys()):
    avg_insurance_costs_women[age] = round(sum(insurance_costs_women[age]) / len(insurance_costs_women[age]), 2)
```

```python

#setting up plot
ax = plt.subplot(1,1,1)
ax.set_xticks([i for i in range(0,50,3)])



#men plot
ages = [age for age in avg_insurance_costs_men]
costs = [cost for cost in avg_insurance_costs_men.values()]
ax.plot(ages,costs, color='brown', label='Men')

#women plot
ages = [age for age in avg_insurance_costs_women]
costs = [cost for cost in avg_insurance_costs_women.values()]
ax.plot(ages, costs, label='Women')


plt.xlabel('Ages')
plt.ylabel('Costs in Dollars')
plt.title('Average Insurance Costs Broken Down by Gender')
plt.legend()

ax.axis([0,45,3000,27000])
plt.show()
```
